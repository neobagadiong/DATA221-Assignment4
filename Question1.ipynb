```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
breastCancerData = load_breast_cancer()
```

```python
#Question 1

feature_matrix =  breastCancerData.data
target_labels = breastCancerData.target

classes, counts = np.unique(target_labels, return_counts=True)
sampleCountPerClass = dict(zip(classes, counts))

print('X shape:',feature_matrix.shape, '\ny shape:',target_labels.shape, '\nClass:Count :',sampleCountPerClass)


'''
- Data set is imbalanced with one class about 50% more smaples than the other.
- Imbalanced datasets lead to overfitting on the class with more samples and underfitting on the smaller sample size.
'''


```

```python
#Question 2

features_train, features_test, labels_train, labels_test = train_test_split(feature_matrix, target_labels, test_size=0.2,random_state=42)

decision_tree_classifier = DecisionTreeClassifier(criterion='entropy')
decision_tree_classifier.fit(features_train, labels_train)

predicted_labels_train = decision_tree_classifier.predict(features_train)
predicted_labels_test = decision_tree_classifier.predict(features_test)

accuracy_train = accuracy_score(labels_train, predicted_labels_train)
accuracy_test = accuracy_score(labels_test, predicted_labels_test)

print('Training Accuracy: ',accuracy_train)
print('Testing Accuracy: ',accuracy_test)

'''
  - Entropy is how mixed the classes of samples are in the current node based on preportion of each class.
    High entropy, closer to 1, means the node contains a variety of classes and they are about even in count.
    Low entropy, closer to or equal to 0, means 1 class is dominant in the samples in the current node or just 1 class is present in the node.

  - high accuracy for training and high accuracy for test shows it shouldn't be over or under fitted because both datasets show similarly high accuracy. 
'''
```

```python
#Question 3
features_trainC, features_testC, labels_trainC, labels_testC = train_test_split(feature_matrix, target_labels, test_size=0.2,random_state=40)

decision_tree_classifier_constrained = DecisionTreeClassifier(criterion='entropy',max_depth=3)
decision_tree_classifier_constrained.fit(features_trainC, labels_trainC)

predicted_labels_trainC = decision_tree_classifier.predict(features_trainC)
predicted_labels_testC = decision_tree_classifier.predict(features_testC)

accuracy_train_constrained = accuracy_score(labels_trainC, predicted_labels_trainC)
accuracy_test_constrained = accuracy_score(labels_testC, predicted_labels_testC)

topFeaturesDF = pd.DataFrame({'Feature':breastCancerData.feature_names,'Importance':decision_tree_classifier_constrained.feature_importances_}).sort_values(by='Importance', ascending=False)[:5]


print('Training Accuracy: ',accuracy_train_constrained)
print('Testing Accuracy: ',accuracy_test_constrained)

print('Top 5 Features:')
print(topFeaturesDF)

```
